```julia
# Kernel maker.
function make_kernel(alpha, rho)
    return alpha^2 * transform(SqExponentialKernel(), 1/(sqrt(2)*rho))
end

# logpdf for GP model.
function gp_logprob(x, y, alpha, rho, sigma)
    kernel = make_kernel(alpha, rho)
    f = GP(kernel)
    fx = f(x, sigma^2)
    return logpdf(fx, y)
end;

# GP model definition in Turing.
@model gp_model(X, y, m_alpha=0.0, s_alpha=0.1, m_rho=0.0, s_rho=1.0,
                m_sigma=0.0, s_sigma=1.0) = begin
    alpha ~ LogNormal(m_alpha, s_alpha)
    rho ~ LogNormal(m_rho, s_rho)
    sigma ~ LogNormal(m_sigma, s_sigma)
    
    log_target = gp_logprob(X, y, alpha, rho, sigma)
    Turing.acclogp!(_varinfo, log_target)
end;
```

```julia
# For reproducibility.
Random.seed!(0)

# Compile
@time _ = sample(gp_model(x, y),  HMC(0.01, 1), 1)

# Run
@time hmc_chain = begin
    burn = 1000  # NOTE: The burn in is also returned. Can't be discarded.
    nsamps = 1000
    iterations = burn + nsamps
    stepsize = 0.01
    nleapfrog = 100
 
    sample(gp_model(x, y), HMC(stepsize, nleapfrog), iterations)
end;

# Collect posterior samples
alpha = vec(group(hmc_chain, :alpha).value.data[end-nsamps:end, :, 1]);
rho = vec(group(hmc_chain, :rho).value.data[end-nsamps:end, :, 1]);
sigma = vec(group(hmc_chain, :sigma).value.data[end-nsamps:end, :, 1]);
hmc_samples = Dict(:alpha => alpha, :rho => rho, :sigma => sigma);
```

```julia
# Plot parameter posteriors.
plt.figure(figsize=(9, 3))

pkeys = collect(keys(hmc_samples))
for i in 1:length(pkeys)
    key = pkeys[i]
    plt.subplot(1, 3, i)
    plt.hist(hmc_samples[key], density=true);
    plt.xlabel(key)
    plt.ylabel("density")
end

plt.tight_layout()
```

```julia
# This funciton returns a function for predicting at new points given parameter values.
function make_gp_predict_fn(Xnew, y, X)
    N = size(X, 1)
    M = size(Xnew, 1)
    Q = N + M
    Z = [Xnew; X]
    Dsq = pairwise(SqEuclidean(), Z, dims=1)
    
    return (alpha, rho, sigma) -> let
        K = KernelFunctions.kappa.(SqExponentialKernel(), Dsq/(2*rho^2)) * alpha^2 + LinearAlgebra.I * sigma^2
        Koo_inv = inv(K[(M+1):end, (M+1):end])
        Knn = K[1:M, 1:M]
        Kno = K[1:M, (M+1):end]
        C = Kno * Koo_inv
        m = C * y
        S = Matrix(LinearAlgebra.Hermitian(Knn - C * Kno'))
        mvn = MvNormal(m, S)
        rand(mvn)
    end
end

gp_predict = make_gp_predict_fn(xgrid[:,:], y, x[:, :])
```

```julia
# Get GP predictions.
@time preds = [gp_predict(alpha[i], rho[i], sigma[i]) for i in ProgressBar(1:nsamps)];
```
